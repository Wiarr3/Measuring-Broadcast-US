```python
import os
import pandas as pd

folder_path = "../resources"
unit_ids_per_file = {}

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        try:
            df = pd.read_csv(os.path.join(folder_path, filename), usecols=["unit_id"])
            unique_ids = set(df["unit_id"].unique())
            unit_ids_per_file[filename] = unique_ids
            print(f"{filename}: {len(unique_ids)} unique unit_id")
        except ValueError:
            print(f"{filename}: brak kolumny 'unit_id' – pomijam")
        except Exception as e:
            print(f"{filename}: błąd – {e}")
```

```python
import os
import pandas as pd
from collections import defaultdict

folder_path = "../resources"
targets_per_unit = defaultdict(set)

# Przeszukujemy pliki
for filename in os.listdir(folder_path):
    if not filename.endswith(".csv"):
        continue
    try:
        df = pd.read_csv(os.path.join(folder_path, filename), usecols=["unit_id", "target"])
        for row in df.itertuples(index=False):
            targets_per_unit[row.unit_id].add(row.target)
    except ValueError:
        continue
    except Exception as e:
        print(f"{filename}: błąd – {e}")
```

```python
target_count_map = {uid: len(targets) for uid, targets in targets_per_unit.items()}
target_count_map
```

```python
import os
import pandas as pd
from collections import defaultdict

folder_path = "../resources"
samknows_targets_per_unit = defaultdict(set)

for filename in os.listdir(folder_path):
    if not filename.endswith(".csv"):
        continue
    try:
        df = pd.read_csv(os.path.join(folder_path, filename), usecols=["unit_id", "target"])
        for row in df.itertuples(index=False):
            if isinstance(row.target, str) and row.target.endswith(".samknows.com"):
                samknows_targets_per_unit[row.unit_id].add(row.target)
    except ValueError:
        continue
    except Exception as e:
        print(f"{filename}: błąd – {e}")
```

```python
samknows_target_list_map = {
    uid: list(targets)
    for uid, targets in samknows_targets_per_unit.items()
}
samknows_target_list_map
```
