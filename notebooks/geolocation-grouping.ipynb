```python
import pandas as pd
import re

df = pd.read_csv("../resources/curr_traceroute.csv")

first_3_hops = df[df["hop"].isin([1, 2, 3])]

def filter_hop_names(x):
    return x.dropna().unique()[[re.match(r"^[A-Za-z]", str(name)) is not None for name in x.dropna().unique()]]

hop_summary = first_3_hops.groupby("unit_id").agg({
    "hop_address": lambda x: x.dropna().unique(),
    "hop_name": filter_hop_names
})

hop_summary = hop_summary[hop_summary["hop_name"].apply(lambda x: len(x) > 0)]

hop_summary["hop_name"]
```

```python
flat_hop_names = [name for sublist in hop_summary["hop_name"] for name in sublist]
flat_hop_names
```

```python
import re

verizon_map = {
    'BSTNMA': 'Boston, MA',
    'WASHDC': 'Washington, DC',
    'NWRKNJ': 'Newark, NJ',
    'CMDNNJ': 'Camden, NJ',
    'NRFLVA': 'Norfolk, VA',
    'PITBPA': 'Pittsburgh, PA',
    'HRBGPA': 'Harrisburg, PA',
    'BFLONY': 'Buffalo, NY',
    'NYCMNY': 'New York, NY',
    'BFLONY': 'Buffalo, NY',
    'PGHKNY': 'Poughkeepsie, NY',
    'ARTNVAFC': 'Arlington, VA',
    'WASHDCDN': 'Washington, DC'
}

def extract_location(hostnames):
    locations = []

    for host in hostnames:
        match = re.search(r'(?P<city>[a-z]+)\.(?P<state>[a-z]{2})\.', host)
        if match:
            location = f"{match.group('city').capitalize()}, {match.group('state').upper()}"
            locations.append(location)
            continue

        for code, city in verizon_map.items():
            if code in host.upper():
                locations.append(city)
                break

    return locations
```

```python
import re

verizon_map = {
    'BSTNMA': 'Boston, MA',
    'WASHDC': 'Washington, DC',
    'NWRKNJ': 'Newark, NJ',
    'CMDNNJ': 'Camden, NJ',
    'NRFLVA': 'Norfolk, VA',
    'PITBPA': 'Pittsburgh, PA',
    'HRBGPA': 'Harrisburg, PA',
    'BFLONY': 'Buffalo, NY',
    'NYCMNY': 'New York, NY',
    'PGHKNY': 'Poughkeepsie, NY',
    'ARTNVAFC': 'Arlington, VA',
    'WASHDCDN': 'Washington, DC'
}

def extract_location(hostname):
    match = re.search(r'(?P<city>[a-z]+)\.(?P<state>[a-z]{2})\.', hostname)
    if match:
        return f"{match.group('city').capitalize()}, {match.group('state').upper()}"

    for code, city in verizon_map.items():
        if code in hostname.upper():
            return city

    return None

def assign_region(row):
    hop = row["hop"]
    host = row["hop_name"]
    if hop in {1} and isinstance(host, str):
        location = extract_location(host)
        if location:
            return location
    return None

df["region"] = df.apply(assign_region, axis=1)
```
